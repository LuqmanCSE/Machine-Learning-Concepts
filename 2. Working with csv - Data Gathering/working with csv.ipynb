# Importing csv data from a server via url

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/google/dspl/master/samples/google/canonical/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

# converters

you can apply python functions to replace all the values of a column. EX: converters={'column_name':function_name}

### Function

```python
def rename(GenderIDDesc):
    if GenderIDDesc == "FEMALE":
        return "Women 🚺"
    elif GenderIDDesc == "MALE":
        return "Man 🚹"
    else:
        return GenderIDDesc
        
```

```python
rename("MALE")
```

```python
pd.read_csv("Health_AnimalBites.csv", converters={"GenderIDDesc":rename})
```

# na_values parameter

pandas by default considers blank values as NaN but sometimes instead of blank the values is -, etc. so we can use 'na_values' to define custom values as NaN. EX: na_values=['-',etc]

```python
pd.read_csv("Health_AnimalBites.csv", na_values=["UNKNOWN"])
```
