```python
a = 2
```

# storing the responce (ex: responce 200, 404, etc) 

```python
response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=f8223b6e16ffdbd7d7ccf3577a2c765a&page={a}")
```

# transforming it into a dataframe

```python
pd.DataFrame(response.json()["results"])
```

## only extracting needed columns

```python
pd.DataFrame(response.json()["results"])[['id','title','adult','popularity','release_date','vote_average','vote_count']]
```

```python
df = pd.DataFrame() # empty dataset
```

### since the url has total 545 pages we have to iterate ovel all those urls(page=1, page =2...)

```python
for i in range(1,546):
    print(i)
    response = requests.get(f"https://api.themoviedb.org/3/movie/top_rated?api_key=f8223b6e16ffdbd7d7ccf3577a2c765a&page={i}")
    tempDf = pd.DataFrame(response.json()["results"])[['id','title','adult','popularity','release_date','vote_average','vote_count']]
    df = pd.concat([df, tempDf])
```
