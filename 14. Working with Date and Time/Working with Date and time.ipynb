```python
import pandas as pd
import numpy as np
```

```python
date = pd.read_csv('../Data sets/orders.csv')
time = pd.read_csv('../Data sets/messages.csv')
```

#### cols are not in dateTIme dtype

```python
date['date'] = pd.to_datetime(date['date'])
```

```python
date.info()
```

## Extract year

```python
date['year'] = date['date'].dt.year
```

```python
date.head()
```

## Extract month

```python
date['month'] = date['date'].dt.month
```

```python
date.head()
```

### If you want name of the month

```python
date['month_name'] = date['date'].dt.month_name()
```

```python
date.head()
```

## Extract day

```python
date['day'] = date['date'].dt.day
date.head()
```

### Extrat day of week; can be usefull for finding weekend

```python
date['day_of_week'] = date['date'].dt.dayofweek
date.head()
```

## extract name of day of the week

```python
date['day_name'] = date['date'].dt.day_name()
date.head()
```

## finding if the day is weekend or not?

```python
date['is_weekend'] = np.where(date['day_name'].isin(['Saturday','Sunday']), 1,0)

date.head()
```

## week of the year

```python
date["week"] = date['date'].dt.isocalendar().week
date.head()
```

### Quarter

```python
date['quarter'] = date['date'].dt.quarter
date.head()
```

### Find sem

```python
date['sem'] = np.where(date['quarter'].isin([1,2]), 1,2)
date.head()
```

# Extract time elapsed between two dates

```python
import datetime

today = datetime.datetime.today()

today
```

### Full elapsed

```python
today - date['date']
```

### Only day

```python
(today - date['date']).dt.days
```

### Months passed

```python
np.round((today - date['date']).dt.days / 30) #assuming each month has 30 days
```

## extract hours, minutes, seconds

```python
time['Hour'] = time['date'].dt.hour
time['Minute'] = time['date'].dt.minute
time['Second'] = time['date'].dt.second

time.sample(10)
```

## Extract only time from date and time

```python
time['time'] = time['date'].dt.time
time.head()
```

## time difference

```python
today - time['date']
```

### In seconds

```python
(today - time['date'])/np.timedelta64(1,'s')
```

```python
(today - time['date']).dt.days*(24*60*60)
```

### in minutes

```python
(today - time['date'])/np.timedelta64(1,'m')
```

```python
(today - time['date']).dt.days*(24*60)
```

### in hours

```python
(today - time['date'])/np.timedelta64(1,'h')
```

```python
(today - time['date']).dt.days*(24)
```
