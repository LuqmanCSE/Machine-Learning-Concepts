```python
import pickle
import numpy as np
```

## Opening each model trained and assigning names

```python
model = pickle.load(open('models/model.pkl','rb'))
ohe_embarked = pickle.load(open('models/ohe_embarked.pkl','rb'))
ohe_sex = pickle.load(open('models/ohe_sex.pkl','rb'))
```

### Assumed User input

0 = pclass
1 = sibsp
2 = parch
3 = fare
4 = age
5 = embarked
6 = embarked
7 = sex

```python
input = np.array([3.0,0.0,0.0,12.0,28,'S','male'], dtype=object).reshape(1,7) #coz without it shape = (6,) and we want (6,1)
```

```python
input.shape
```

```python
input_sex = ohe_sex.transform(input[:,-1].reshape(1,1)) 
```

```python
input_sex
```

```python
input_embarked = ohe_embarked.transform(input[:,-2].reshape(1,1))
```

```python
input_embarked
```

## add all healthy cols to one df

```python
input_healthy = input[:,0:4]
```

```python
input_healthy.reshape(4,)
```

```python
input_age = input[:,4] #make age col 
```

```python
input_age
```

## Concatenate everything to gether 

```python
### Before concatenating confirm if all are in same dimension
```

```python
print(input_healthy.shape)
print(input_age.shape)
print(input_embarked.shape)
print(input_sex.shape)
```

```python
input_age = input_age.reshape(1,1) #now its matching dimen with others
```

```python
input_transformed = np.concatenate((input_healthy,input_age,input_embarked,input_sex),axis=1)
```

```python
model.predict(input_transformed)
```
