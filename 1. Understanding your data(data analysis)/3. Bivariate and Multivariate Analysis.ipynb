# 3. Boxplot

```python
sns.boxplot(x=titanic['Sex'], y=titanic['Age'], hue=titanic['Survived'])
```

# 4. Distplot (Numerical - Categorical)

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'], hist=False)
sns.distplot(titanic[titanic['Survived']==1]['Age'], hist=False)
```

# 5. Heatmap (Categorical - Categorical)

*Used to check relation between two categorical column*

```python
pd.crosstab(titanic['Pclass'], titanic['Survived'])
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'], titanic['Survived']))
```

```python
titanic.groupby('Sex')['Survived'].mean()*100  #Gives the % of people survived from each class(or any other factor)
```

```python
(titanic.groupby('Sex')['Survived'].mean()*100).plot(kind='bar')
```

# 6. Cluster map (categorical - categorical)

```python
pd.crosstab(titanic['SibSp'], titanic['Survived'])
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'], titanic['Survived']))
```

*This shows how close values are to each other, like: people traveling with 5,4,6, or 3 partners were are in similar situation(only few survived)*
